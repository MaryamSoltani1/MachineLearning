to categorical

```python
# Convert y_train into one-hot format
from tensorflow.keras.utils import to_categorical
temp = []
for i in range(len(train_labels)):

  temp.append(to_categorical(train_labels[i], num_classes=10))
train_labels = np.array(temp)
# Convert y_test into one-hot format
temp = []
for i in range(len(test_labels)):    
    temp.append(to_categorical(test_labels[i], num_classes=10))
test_labels = np.array(temp)
```

# **set up the layers**

```python
import tensorflow as tf
import keras
from keras.layers import Dense, Flatten
from keras.models import Sequential
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(8, activation='relu'))
model.add(Dense(7, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
#model.compile(optimizer='adam',
   #           loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
   #           metrics=['accuracy'])
model.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=['acc'])
```

```python
model.build((60000, 28, 28))
model.summary()
```

```python
print(train_labels.shape)
print(train_images.shape)
```

```python
train_labels[0]
```

```python
#model.fit(train_images, train_labels, epochs=10)
history=model.fit((train_images), train_labels,validation_data=(test_images, test_labels), epochs=50) 
        #  validation_data=(test_images, test_labels))
```

```python
predictions = model.predict(test_images)
print(predictions)
```

```python
predictions = np.argmax(predictions, axis=1)
print(predictions)
```
