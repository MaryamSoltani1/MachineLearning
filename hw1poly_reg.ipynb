```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import io
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
dataset = pd.read_csv(io.BytesIO(uploaded['Dataset1.csv']))
x = dataset['x']
y = dataset['y']

#x=sc.fit_transform(x.values.reshape(-1, 1) )
#y=sc.fit_transform(y.values.reshape(-1, 1) )
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
#min_max_scaler = preprocessing.MinMaxScaler()
#x_scaled = min_max_scaler.fit_transform(x.values.reshape(-1,1))
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
x_train = x_train.values
x_test = x_test.values

y_train = y_train.values
y_test = y_test.values
x
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score


linear = np.poly1d(np.polyfit(x_train, y_train, 1))



y_predicted_train = np.ones(len(x_train))
y_predicted_test = np.ones(len(x_test))

for i in range(0, len(x_train)):
  y_predicted_train[i] = linear(x_train[i])

for i in range(0, len(x_test)):
  y_predicted_test[i] = linear(x_test[i])

myline = np.linspace(-200, 1000, 10)
plt.scatter(x,y,label='Original Data')
plt.plot(myline, linear(myline), 'r', label='Linear Regression')
plt.legend()
plt.title('Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show

RMSE_train = mean_squared_error(y_train, y_predicted_train, squared=False)
MAE_train = mean_absolute_error(y_train, y_predicted_train)
r2_score_train = r2_score(y_train, y_predicted_train)

RMSE_test = mean_squared_error(y_test, y_predicted_test, squared=False)
MAE_test = mean_absolute_error(y_test, y_predicted_test)
r2_score_test = r2_score(y_test, y_predicted_test)

print('RMSE_train = ', RMSE_train)
print('MAE_train = ', MAE_train)
print('r2_score_train = ', r2_score_train)
print('RMSE_test = ', RMSE_test)
print('MAE_test = ', MAE_test)
print('r2_score_test = ', r2_score_test)
```
