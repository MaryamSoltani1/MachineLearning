```python
from sklearn.model_selection import train_test_split
#X = df.drop('E/P',axis=1)
#y = df[['E/P']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state=42)

from sklearn.tree import DecisionTreeClassifier
clf_model = DecisionTreeClassifier(criterion="entropy", random_state=42,max_depth=6)
clf_model.fit(x_train,y_train)
y_predict = clf_model.predict(x_test)

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_test,y_predict)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=441)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state=249)

scores = []
k_range = range(1,13)
for k in k_range:
  tree_clf = DecisionTreeClassifier(criterion='entropy',max_depth=k)
  tree_clf.fit(x_train, y_train)
  y_pred_val = tree_clf.predict(x_val)
  scores.append(accuracy_score(y_val, y_pred_val))
plt.plot(k_range,scores)
plt.xlabel('Maximum depth of tree')
plt.ylabel('Accuracy on validation set')
plt.show
```
