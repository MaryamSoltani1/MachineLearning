```python
def remove_URL(text):
    text = str(text)
    url = re.compile(r"https?://\S+|www\.\S+")
    return url.sub(r"", text)
def remove_HTML(text):
    text = str(text)
    html = re.compile(r"<.*?>")
    return html.sub(r"", text)
def remove_emoji(text):
    text = str(text)
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE,
    )
    return emoji_pattern.sub(r"", text)
def normalize_text(text):
    return normalizer.normalize(text)
def remove_number(text):
    text = str(text)
    result = re.sub(r'\d+','', text)
    return result
def clean_up_pipeline(data):
    cleaning_functions = [
        str,
        str.strip,
        str.lower,
        remove_number,
        remove_URL,
        remove_HTML,
        remove_emoji,
        ]
    for function in cleaning_functions:
        df.loc[:, text_column] = df[text_column].map(function)
    return df
```

```python
puncs = ['،','.',',',':',';','"',"'",'/','\\','_','-']
def load_stopwords():
    f = open("/content/drive/MyDrive/stopwords.txt", "r", encoding='utf8')
    stopwords = f.read()
    stopwords = stopwords.split('\n')
    stopwords = set(stopwords)
    custom_stop_words = {'آنكه','آيا','بدين','براين','بنابر','میشه','میکنه','باشه','سلام','میکشه','اونی','و','در','از','چه'}
    stopwords = stopwords  | custom_stop_words
    stopwords = list(stopwords)[1:]
    unwanted_num = {'خوش','بهتر','بد','خوب','نیستم','عالی','نیست','فوق','بهترین','خیلی', 'نبود'} 
    stopwords = [ele for ele in stopwords if ele not in unwanted_num]
    return stopwords
def clean_doc(doc):
    doc = normalizer.normalize(doc)
    tokenized = word_tokenize(doc)
    tokens = []
    for t in tokenized:
      temp = t
      for p in puncs:
        temp = temp.replace(p, '')
      tokens.append(temp)
    stop_set = load_stopwords()
    tokens = [w for w in tokens if not w in stop_set]
    tokens = [w for w in tokens if not len(w) <= 1]
    tokens = [w for w in tokens if not w.isdigit()]
    tokens =  [w for w in tokens if not re.match(r'[A-Z]+', w, re.I)]
    tokens = ' '.join(tokens)
    return tokens
```
